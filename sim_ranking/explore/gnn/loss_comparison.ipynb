```python
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.ndimage import gaussian_filter1d

import sim_ranking as sr
import ml_tools as mlt
```

```python
base_result_dir = Path("/Users/claudy/dev/work/data/sim_ranking/results/gnn")

gnn_result_dir_names_id = {
    "0411_0812_cv_v4p3NZGMDB_latest_noClamp_3e5s": "no-clamp",
    "0411_0922_cv_v4p3NZGMDB_latest_10Clamp_3e5s": "10-clamp",
    "0411_0950_cv_v4p3NZGMDB_latest_5Clamp_3e5s": "5-clamp",
    "0411_1024_cv_v4p3NZGMDB_latest_3Clamp_3e5s": "3-clamp",
    "0411_1047_cv_v4p3NZGMDB_latest_4Clamp_3e5s": "4-clamp",
    "0411_1111_cv_v4p3NZGMDB_latest_3p5Clamp_3e5s": "3.5-clamp",
}
```

```python
metric_dict = {}
for cur_dir, cur_id in gnn_result_dir_names_id.items():
    cur_metrics = pd.read_pickle(base_result_dir / cur_dir / "metrics.pickle")
    metric_dict[cur_id] = cur_metrics
```

### Mean CV Loss

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

for cur_id in metric_dict.keys():
    cur_metrics = metric_dict[cur_id]

    x = np.arange(cur_metrics.shape[0])
    cur_train_mean = cur_metrics.sel[:, :,"w_loss_hist_train"].mean(axis=1)
    cur_val_mean = cur_metrics.sel[:, :,"w_loss_hist_val"].mean(axis=1)

    
    ax1.plot(x, cur_train_mean, label=cur_id)
    ax2.plot(x, cur_val_mean, label=cur_id)

ax1.set_title("Train Loss")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(linewidth=0.5, alpha=0.5, linestyle="--")
ax1.legend()
ax1.set_ylim(-3, -2)
ax1.set_xlim(0, 250)

ax2.set_title("Validation Loss")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.grid(linewidth=0.5, alpha=0.5, linestyle="--")
ax2.set_ylim(-3, -2)
ax2.set_xlim(0, 250)

fig.tight_layout()
```

### Standard Deviation Of CV Loss

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

for cur_id in metric_dict.keys():
    cur_metrics = metric_dict[cur_id]

    x = np.arange(cur_metrics.shape[0])
    cur_train_std = cur_metrics.sel[:, :,"w_loss_hist_train"].std(axis=1)
    cur_val_std = cur_metrics.sel[:, :,"w_loss_hist_val"].std(axis=1)

    
    ax1.plot(x, cur_train_std, label=cur_id)
    ax2.plot(x, cur_val_std, label=cur_id)

ax1.set_title("Train Standard Deviation")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(linewidth=0.5, alpha=0.5, linestyle="--")
ax1.legend()
# ax1.set_ylim(-3, -2)
ax1.set_xlim(0, 250)

ax2.set_title("Validation Standard Deviation")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.grid(linewidth=0.5, alpha=0.5, linestyle="--")
# ax2.set_ylim(-3, -2)
ax2.set_xlim(0, 250)

fig.tight_layout()
```

### Smoothed Mean & Standard Deviation

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

for cur_id in metric_dict.keys():
    cur_metrics = metric_dict[cur_id]

    x = np.arange(cur_metrics.shape[0])
    cur_train_mean = cur_metrics.sel[:, :,"w_loss_hist_train"].mean(axis=1)
    cur_val_mean = cur_metrics.sel[:, :,"w_loss_hist_val"].mean(axis=1)

    cur_train_mean = gaussian_filter1d(cur_train_mean, sigma=5)
    cur_val_mean = gaussian_filter1d(cur_val_mean, sigma=5)
    
    ax1.plot(x, cur_train_mean, label=cur_id)
    ax2.plot(x, cur_val_mean, label=cur_id)

ax1.set_title("Train Loss")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss")
ax1.grid(linewidth=0.5, alpha=0.5, linestyle="--")
ax1.legend()
ax1.set_ylim(-3, -2)
ax1.set_xlim(0, 250)

ax2.set_title("Validation Loss")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.grid(linewidth=0.5, alpha=0.5, linestyle="--")
ax2.set_ylim(-3, -2)
ax2.set_xlim(0, 250)

fig.tight_layout()
```
