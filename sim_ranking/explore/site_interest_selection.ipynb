### Site of Interest Selection
Explores the effect of site of interest selection criteria.   
Used to be just all sites available with at least 1 site observations within the maximum separation distance.  
However, many of these sites experience such small GM that they are not of interest. Therefore, as advised by Brendon, this has been updated to all sites that experience at least PGA 0.1g based on the BR13 GMP

```python
old_data_dir = Path("/Users/claudy/dev/work/data/sim_ranking/results/ml/archive/0529_0850_lb_15_100_SCweighted_probL2_0p5")
new_data_dir = Path("/Users/claudy/dev/work/data/sim_ranking/results/ml/0618_0811_lb_15_100_SCweighted_probL2_0p5")
# new_data_dir = Path("/Users/claudy/dev/work/data/sim_ranking/results/ml/0618_0816_lb_30_100_SCweighted_probL2_0p5")
```

```python
old_val_df = pd.read_parquet(old_data_dir / "val_scenario_summary.parquet")
old_train_df = pd.read_parquet(old_data_dir / "train_scenario_summary.parquet")
old_df = pd.concat([old_val_df, old_train_df])

new_val_df = pd.read_parquet(new_data_dir / "val_scenario_summary.parquet")
new_train_df = pd.read_parquet(new_data_dir / "train_scenario_summary.parquet")
new_df = pd.concat([new_val_df, new_train_df])
```
